## Análise de Componentes Principais (PCA - Principal Component Analysis)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

x = df.select_dtypes(include='number')

# 3. Normalizar os dados (PCA é sensível à escala)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# 4. Aplicar PCA
pca = PCA(n_components=2)  # Reduzindo para 2 dimensões
x_pca = pca.fit_transform(x_scaled)

# 5. Visualizar resultado
plt.scatter(x_pca[:, 0], x_pca[:, 1])
plt.title("PCA - Redução para 2 Componentes")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.grid()
plt.show()

# 6. Explicar variância preservada
print("Variância explicada por cada componente:", pca.explained_variance_ratio_)
print("Variância total preservada:", sum(pca.explained_variance_ratio_))
```

```python
#conjunto de dados numéricos brutos
x = df.select_dtypes(include='number')

#padronizar os dados
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

#aplicando PCA
pca = PCA()
x_pca = pca.fit_transform(x_scaled)

#entender a variância explicada
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_.cumsum(), marker='o')
plt.title('Variância Explicada Acumulada pelos Componentes da PCA')
plt.xlabel('Número de Componentes')
plt.ylabel('Variância Acumulada')
plt.grid(True)
plt.show()

#selecionar os principais componentes
pca = PCA(n_components=0.95)  # manter 95% da variância explicada
x_reduced = pca.fit_transform(x_scaled)
```

## Reduzir para 95% da Variância e Interpretar os Componentes

```python
# Manter o número de componentes necessários para preservar 95% da variância explicada
pca_95 = PCA(n_components=0.95)
x_reduced = pca_95.fit_transform(x_scaled)

# Loadings da PCA (como as variáveis originais contribuem para cada componente)
loadings = pd.DataFrame(pca_95.components_, columns=x.columns)
print("PCA Loadings (Top Components):")
print(loadings.head())
```

## Visualizar PCA em 2D

```python
# Criar um DataFrame com os primeiros 2 componentes principais
pca_df = pd.DataFrame(x_reduced[:, :2], columns=['PC1', 'PC2'])

# Target/label 
pca_df['Premium_Amount'] = df['Premium_Amount']

sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue = 'Premium_Amount')  
plt.title('2D PCA Projection')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Regressão Linear com PCA

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split

x = df.select_dtypes(include='number').drop(columns=['Premium_Amount'])  
y = df['Premium_Amount']  # target

# Scale 
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Aplicar PCA para reter 95% da variância
pca = PCA(n_components=0.95)
x_pca = pca.fit_transform(x_scaled)

# Dividir o conjunto de dados em treino e teste
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=42)

# Treinar o modelo de Regressão Linear
model = LinearRegression()
model.fit(x_train, y_train)

# Prever e avaliar
y_pred = model.predict(x_test)
r2 = r2_score(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print(f"R² score: {r2:.3f}")
print(f"RMSE: {rmse:.3f}")
```

## Interpretar os coeficientes do modelo

```python
coefficients = model.coef_
for i, coef in enumerate(coefficients):
    print(f"PC{i+1}: {coef:.4f}")
```

## Visualizar Resíduos e Erros (modelo baseado em PCA)

```python
# Resíduos
residuals = y_test - y_pred

# Plotando resíduos versus valores previstos
plt.figure(figsize=(8, 5))
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Valores Previstos")
plt.ylabel("Resíduos (Real - Previsto)")
plt.title("Gráfico Residual (Regressão Linear Baseada em PCA)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Construir e Comparar com o Modelo Usando as Características Originais

```python
# Modelo sem PCA
# Divisão de treino/teste nas características originais e escalonadas
x_scaled_orig = scaler.fit_transform(x)  
x_train_orig, x_test_orig, y_train_orig, y_test_orig = train_test_split(x_scaled_orig, y, test_size=0.2, random_state=42)

# Regressão linear nas características originais
model_orig = LinearRegression()
model_orig.fit(x_train_orig, y_train_orig)

# Prever e avaliar
y_pred_orig = model_orig.predict(x_test_orig)
r2_orig = r2_score(y_test_orig, y_pred_orig)
rmse_orig = mean_squared_error(y_test_orig, y_pred_orig, squared=False)

print(f"\n Modelo sem PCA:")
print(f"R² score: {r2_orig:.3f}")
print(f"RMSE: {rmse_orig:.3f}")
```

## Comparar Ambos os Modelos

```python
print("\n Resumo da Comparação dos Modelos:")
print(f"{'':<20} {'com PCA':<15} {'sem PCA':<15}")
print(f"{'R² Score':<20} {r2:.3f}<15 {r2_orig:.3f}<15")
print(f"{'RMSE':<20} {rmse:.3f}<15 {rmse_orig:.3f}<15")
```

## Visualizar Ambos os Resíduos

```python
plt.figure(figsize=(14, 5))

# Resíduos do modelo PCA
plt.subplot(1, 2, 1)
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residuals: PCA Model")
plt.xlabel("Predicted")
plt.ylabel("Residuals")

# Resíduos do modelo original
residuals_orig = y_test_orig - y_pred_orig
plt.subplot(1, 2, 2)
sns.scatterplot(x=y_pred_orig, y=residuals_orig)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residuals: Original Features")
plt.xlabel("Predicted")
plt.ylabel("Residuals")

plt.tight_layout()
plt.show()
```
